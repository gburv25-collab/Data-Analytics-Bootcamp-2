```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data
df = pd.read_csv("/content/Retail_Sales_2024.csv", parse_dates=["Date"])

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Date"].dtype

# 2. Create Revenue
df["Revenue"] = df["Quantity"] * df["Price"] * (1 - df["Discount_Rate"])

```

```python
# Question 1: Which product categories perform best?

cat_rev = df.groupby("Category")["Revenue"].sum().sort_values()

plt.figure()
cat_rev.plot(kind="bar")
plt.title("Total Revenue by Category")
plt.ylabel("Revenue")
plt.xlabel("Category")
plt.tight_layout()
plt.show()

# Answer: Which bar is highest? How big is the gap between top and bottom category?
```

```python
# Question 2: How do monthly sales patterns change?

# Create a monthly series
monthly_rev = df.set_index("Date").resample("ME")["Revenue"].sum()

plt.figure()
monthly_rev.plot()
plt.title("Monthly Revenue Trend â€“ 2024")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.tight_layout()

# Y-axis to start at zero
plt.ylim(bottom=0)
plt.ylim(0, 60000)

plt.tight_layout()
plt.show()

# Answer: Discussion: Identify peak months. How does this compare to their expectations of retail seasonality?
```

```python
# Stretch: Relationship between units sold, price, and revenue

plt.figure()
plt.scatter(df["Quantity"], df["Price"])
plt.title("Price vs Quantity Sold")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.tight_layout()
plt.show()

# Extension:

# Answer: Do you see any patterns or clusters? Where are the high-revenue points likely to be on this scatter?
```
